```python
import dateutil
import datetime
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
#n=int(input("Enter the number of stocks :"))

lst=[] #For storing the assets


tick=input("Enter the stock name listed in NSE : ")
tick=tick+'.NS'
lst.append(tick)
    
lst.append('^NSEI')

print(lst)

# Set the start and end date
start_date = '2016-01-01' #Beta is typically calculated on past 5 yrs data
end_date = '2021-08-03'
df=[]
for i in range(len(lst)):
    ticker=lst[i]
    data=yf.download(ticker, start_date, end_date)
    df.append(data)
```

```python
stock_data = pd.concat(df, axis=1, ignore_index=False)
stock_data.head()
```

```python
pf_data=stock_data['Adj Close']
pf_data.head()

pf_data.columns=lst
pf_data.tail()
```

```python
log_returns = np.log(pf_data / pf_data.shift(1))
log_returns
```

```python
cov = log_returns.cov() * 250
cov
```

```python
cov_with_market = cov.iloc[0,1]
cov_with_market
```

```python
market_var = log_returns['^NSEI'].var() * 250
market_var
```

# Beta Calculation

```python
STOCK_beta = cov_with_market / market_var
STOCK_beta
```

# Expected return of the stock(CAPM)

```python
STOCK_er = 0.0675 + STOCK_beta * 0.05
STOCK_er
```

# Sharpe Ratio

```python
print (tick)
Sharpe = (STOCK_er - 0.0675) / (log_returns[tick].std() * 250 ** 0.5)
Sharpe
```
