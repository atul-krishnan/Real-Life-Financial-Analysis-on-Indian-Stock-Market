```python
import dateutil
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import yfinance as yf
```

```python
n=int(input("Enter the number of stocks :"))

lst=[]
for i in range(n):
    tick=input("Enter the stock name listed in NSE : ")
    t=tick+'.NS'
    lst.append(t)
    
print(lst)

# Set the start and end date
start_date = '2000-01-01'
end_date = '2021-12-21'
df=[]
for i in range(len(lst)):
    ticker=lst[i]
    data=yf.download(ticker, start_date, end_date)
    df.append(data)
    
#print(df)
    
# # Set the ticker
# ticker = 'INFY.NS'

# # Get the data
# data = yf.download(ticker, start_date, end_date)

# # Print 5 rows
# data.tail()
```

```python
stock_data = pd.concat(df, axis=1, ignore_index=False)
```

```python
stock_data1=stock_data['Adj Close']
stock_data1.columns=lst
stock_data1.tail()
```

```python
stock_data1.iloc[0]
```

```python
plt.style.use('bmh')
(stock_data1 / stock_data1.iloc[0] * 100).plot(figsize = (15, 6));
plt.show()
```

```python
sec_returns = np.log(stock_data1 / stock_data1.shift(1))
```
